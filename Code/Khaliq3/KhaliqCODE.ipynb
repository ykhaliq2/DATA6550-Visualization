```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'train.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info()

# Display the first few rows of the dataset
df.head()
```

## GOOD VISUALIZATIONS

```python
# 1. Survival Rate by Embarked Location
plt.figure(figsize=(8, 6))
sns.barplot(x='Embarked', y='Survived', data=df, ci=None, palette='cool')
plt.title('Survival Rate by Embarked Location')
plt.ylabel('Survival Rate')
plt.show()
```

Survival Rate by Embarked Location:

- Passengers who embarked from Cherbourg (C) had the highest survival rate, followed by Queenstown (Q) and Southampton (S).

```python
# 2. Survival Rate by Family Size
df['FamilySize'] = df['SibSp'] + df['Parch']
plt.figure(figsize=(10, 6))
sns.barplot(x='FamilySize', y='Survived', data=df, ci=None, palette='crest')
plt.title('Survival Rate by Family Size')
plt.ylabel('Survival Rate')
plt.show()
```

Survival Rate by Family Size:

- Smaller family sizes (1-3 members) had higher survival rates compared to larger families.
- Solo travelers (0 family size) had a lower survival rate than small families but higher than large families.

## BAD VISUALIZATIONS

```python
# 1. Bad Visualization: Pie Chart with Too Many Categories and Poor Color Choice
plt.figure(figsize=(8, 8))
plt.pie(df['Pclass'], labels=df['PassengerId'], autopct='%1.1f%%', colors=['#FF0000', '#00FF00', '#0000FF'])
plt.title('Horrible Pie Chart: Passenger Class by Passenger ID')
plt.show()

# 2. Bad Visualization: Overlapping Scatter Plot with Nonsensical Axes
plt.figure(figsize=(10, 6))
plt.scatter(df['PassengerId'], df['Name'].str.len(), c='red', alpha=0.7)
plt.title('Nonsensical Scatter Plot: Passenger ID vs Name Length')
plt.xlabel('Passenger ID')
plt.ylabel('Name Length')
plt.show()
```

Horrible Pie Chart:

A pie chart with too many categories (one for each passenger) and poor color choices, making it impossible to interpret.
Labels are overlapping and illegible.
This chart is overly complicated and misleading as it tries to compare too many groups.

Nonsensical Scatter Plot:

A scatter plot comparing Passenger ID with Name Length, which has no logical relationship.
The axes are arbitrary and do not provide any meaningful insight.
It creates visual clutter with overlapping points and no clear trend or pattern.
